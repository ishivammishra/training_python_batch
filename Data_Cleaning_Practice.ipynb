```python
extra_space_names = df[df['Name'].str.startswith(' ') | df['Name'].str.endswith(' ')]
extra_space_names
```

```python
print('Total names with extra spaces =', len(extra_space_names))
```

```python
import re

def has_special_char(name):
    if pd.isnull(name):
        return False
    return bool(re.search(r'[^a-zA-Z\s]', name))

special_char_names = df[df['Name'].apply(has_special_char)]
special_char_names.head(20)
```

```python
print('Total names with extra spaces =', len(special_char_names))
```
