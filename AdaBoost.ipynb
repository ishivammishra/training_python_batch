```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score , precision_score , recall_score  , f1_score
```

```python
#make classification is a generator function which return both the input and output thats 
X,y = make_classification(n_samples=1000 , n_features=5 , n_classes=2 , random_state=42)
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3 , random_state=42)
```

```python
basemod = DecisionTreeClassifier(max_depth=1)
```

```python
adamod = AdaBoostClassifier(estimator=basemod , n_estimators =200 , random_state=42)
```

```python
adamod.fit(X_train , y_train)
```

```python
yadapred = adamod.predict(X_test)
```

```python
print('Training Score : ', adamod.score(X_train , y_train))
print('Test Score : ', adamod.score(X_test , y_test))
```

```python
print('Accuracy is :', accuracy_score(y_test ,yadapred ))
print('F1 is :', f1_score(y_test ,yadapred ))
print('Precision is :', precision_score(y_test ,yadapred ))
print('Recall is :', recall_score(y_test ,yadapred ))
```
